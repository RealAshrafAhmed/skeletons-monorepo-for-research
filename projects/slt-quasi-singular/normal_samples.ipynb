# Import Required Libraries
Import libraries such as numpy, matplotlib, and pandas for data generation, visualization, and saving.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Define Parameters for Normal Distribution
Set the mean, standard deviation, and sample size for the normal distribution.

```python
mean = 0.0
std_dev = 1.0
sample_size = 200
```

# Generate Normal Samples
Use numpy to generate samples from a normal distribution with the specified parameters.

```python
np.random.seed(42)
samples = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
samples[:10]  # Show first 10 samples
```

# Visualize the Samples
Plot a histogram and density plot of the generated samples using matplotlib and seaborn.

```python
plt.figure(figsize=(10, 6))
sns.histplot(samples, kde=True, bins=30, color='skyblue')
plt.title('Histogram and Density of Standard Normal Samples')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

# Save Samples to File
Save the generated samples to a CSV file using pandas for further analysis or sharing.

```python
df = pd.DataFrame({'samples': samples})
df.to_csv('normal_samples.csv', index=False)
df.head()
```

# Save Samples to Data Folder
Save the generated samples to the shared data directory for use in other notebooks.

```python
import os
os.makedirs("../../data", exist_ok=True)
df.to_csv("../../data/normal_samples.csv", index=False)
print("Samples saved to ../../data/normal_samples.csv")
```
